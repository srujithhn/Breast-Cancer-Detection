 # Loading Dataset

```python
import pandas as p
import numpy as np

data=p.read_csv('brcal.csv')
data
```

# Cleaning and Feature selection

```python
data.info()
data.describe()
data.dropna()
data.isnull().sum()
```

```python
data.drop(["id"],axis=1,inplace=True)
data
```

```python
data.diagnosis.value_counts()
```

 #  Feature selection 

```python
from sklearn.model_selection import train_test_split
diag_map={'M':1,'B':0}
data['diagnosis']=data['diagnosis'].map(diag_map)
```

```python
x = data.iloc[:, 1:]  
x = x.drop(columns=["Unnamed: 32"])
y=data[['diagnosis']]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=30)
```

 # Model - KNN 

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier()
knn.fit(x_train,y_train)
knn_pred=knn.predict(x_test)

# warnings removing
import warnings
warnings.filterwarnings('ignore', category=DeprecationWarning)
```

 # Accuracy score and confusion matrix for KNN

```python
from sklearn.metrics import confusion_matrix , accuracy_score
cm=confusion_matrix(knn_pred,y_test)
print(np.array(cm))
print('K Neighbors Classifier accuracy score is ',accuracy_score(knn_pred,y_test))
```

 # Model - Logistic regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
lr.fit(x_train,y_train)
lr_pred=lr.predict(x_test)

# warnings removing
import warnings
warnings.filterwarnings('ignore', category=DeprecationWarning)
```

 # Accuracy score and confusion matrix for Logistic regression

```python
cm=confusion_matrix(lr_pred,y_test)
print(np.array(cm))
print('Logistic regression accuracy score is ',accuracy_score(lr_pred,y_test))
```

 # Model - Gnn(Naive Bayes)

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(x_train, y_train)
gnb_pred=gnb.predict(x_test)
```

 # Accuracy score and confusion matrix for Naive Bayes

```python
cm=confusion_matrix(gnb_pred,y_test)
print(np.array(cm))
print('Naive Bayes accuracy score is ',accuracy_score(gnb_pred,y_test))
```

 # K cross validation 

```python
from sklearn.model_selection import cross_val_score
```

```python
acc_all=[]
csv_all=[]
score=cross_val_score(knn,x_train,y_train,cv=100)
acc_all.append(accuracy_score(knn_pred,y_test))
csv_all.append(np.mean(score))
print('Accuracy of KNN',acc_all)
print('K cross validation scores  of KNN',csv_all)
# warnings removing
import warnings
warnings.filterwarnings('ignore', category=DeprecationWarning)
```

```python
acc_all=[]
csv_all=[]
score=cross_val_score(lr,x_train,y_train,cv=10)
acc_all.append(accuracy_score(lr_pred,y_test))
csv_all.append(np.mean(score))
print('Accuracy of Logistic regression',acc_all)
print('K cross validation scores  of Logistic regression',csv_all)

# warnings removing
import warnings
warnings.filterwarnings('ignore', category=DeprecationWarning)
```

```python
acc_all=[]
csv_all=[]
score=cross_val_score(gnb,x_train,y_train,cv=10)
acc_all.append(accuracy_score(gnb_pred,y_test))
csv_all.append(np.mean(score))
print('Accuracy of Naive Bayes',acc_all)
print('K cross validation scores  of Naive Bayes',csv_all)
```
